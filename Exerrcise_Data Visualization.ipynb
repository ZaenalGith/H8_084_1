```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', 500)
```

```python
df = pd.read_csv('melb_data.csv')
df.head()
```

Nomor 1 : Histogram
    
Buatlah Histogram untuk menampilkan distribusi data harga rumah (kolom “Price”)!

```python
df.Price.plot(kind='hist', figsize = (8, 5), bins = 200)

plt.title('Distribusi Harga Rumah', size = 15)   
plt.ylabel('Frequency')
plt.xlabel('Harga Rumah')
```

```python
sns.histplot(df.Price)

plt.title('Distribusi Harga Rumah', size = 15)   
plt.ylabel('Frequency')
plt.xlabel('Harga Rumah')
```

```python
plt.style.use('seaborn')
sns.histplot(df['Price'])
plt.title('DIstribusi Harga Rumah', size = 15)
plt.show()
```

Nomor 2 : Box Plot
    
Buatlah Box Plot untuk menampilkan distribusi data jarak ke CBD (kolom “Distance”)! Buatlah Box Plot “Distance” untuk tiap nama region (kolom “Regionname”)!

```python
plt.figure(figsize = (8, 6))
sns.boxplot(x = df['Distance'], y = df["Regionname"])

plt.title('Distribusi ')
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x = df['Distance'], y=df['Regionname'])
plt.title('Distribusi Jarak ke CBD berdasarkan Nama Region', size = 15)
plt.show()
```

Nomor 3 : Bar Plot

Buatlah Bar Plot (diagram batang) untuk membandingkan nilai median harga rumah (kolom “Price”) di tiap tipe rumah (kolom “Type”)!

```python
plt.figure(figsize=(8,6))
sns.barplot(x='Type', y='Price', data=df, estimator=np.median)
plt.title('Perbandingan Nilai Median Harga Rumah', size=15)
plt.show()
```

Nomor 4 : Scatter Plot

Buatlah Scatter Plot (diagram titik) untuk menampilkan korelasi antara harga rumah (kolom “Price”) dengan Jarak ke CBD (kolom “Distance”)! Bedakan warna titik berdasarkan tipe rumah (kolom “Type”)!

```python
sns.scatterplot(data=df, x='Price', y='Distance', hue='Type')
plt.title('Korelasi Harga Rumah dan Jarak Rumah ke CBD', size=15)
plt.show()
```

Nomor 5 : Line Plot

Buatlah line plot (diagram plot) untuk menampilkan tren perkembangan harga rumah (Kolom "Price") tiap bulan

```python
df['Date'] = pd.to_datetime(df['Date'])

df['Month'] = pd.DatetimeIndex(df['Date']).month
```

```python
plt.figure(figsize=(8,6))
sns.lineplot(x=df.Month,y=df.Price,hue=df.Type)

plt.title('Tren Harga RUmah per BUlan berdasarkan Tipe Rumah', size=15)
```
